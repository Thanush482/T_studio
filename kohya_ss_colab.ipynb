## Accessing Kohya SS WebUI via ngrok

Since Gradio's share link creation failed, we can use `ngrok` to create a secure tunnel to your locally running Kohya SS WebUI. This will give you a public URL to access the interface.

````python
print('Installing ngrok...')
!pip install pyngrok
from IPython.display import clear_output
clear_output()
print('ngrok installed successfully!')
````

### ngrok Authentication

To use ngrok, you need an authentication token.
1. Go to [https://dashboard.ngrok.com/signup](https://dashboard.ngrok.com/signup) and create an account.
2. After signing up, get your Authtoken from [https://dashboard.ngrok.com/get-started/your-authtoken](https://dashboard.ngrok.com/get-started/your-authtoken).
3. In Colab, add this token to the secrets manager under the "🔑" icon in the left panel. Name the secret `NGROK_AUTH_TOKEN`.

Once you've set the secret, run the next cell to authenticate ngrok.

````python
# Used to securely store your API key
from google.colab import userdata
from pyngrok import ngrok

print('Authenticating ngrok...')
NGROK_AUTH_TOKEN = userdata.get('NGROK_AUTH_TOKEN')
if NGROK_AUTH_TOKEN is None:
    print("Please add your NGROK_AUTH_TOKEN to Colab secrets. Refer to the markdown cell above.")
else:
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    print('ngrok authenticated successfully!')
````

### Restart Kohya SS WebUI and Start ngrok Tunnel

Now, we'll restart the Kohya SS WebUI without the `--share` flag and then use `ngrok` to create a public URL for it. The Kohya SS WebUI runs on port `7860`.

````python
import sys
import os
import threading
import time
from pyngrok import ngrok

# Ensure User and Password are available from previous cell execution
# If you restarted the runtime, you might need to re-run cell `IzS3hvuTtTqW` up to the `User` and `Password` definitions

# These variables should be defined if the previous cell `IzS3hvuTtTqW` was run successfully (before being stopped).
# User = "Thanush123" # Example, replace with actual if needed or rely on kernel state
# Password = "thanush12345" # Example, replace with actual if needed or rely on kernel state

print("Restarting Kohya SS WebUI (without Gradio share) and starting ngrok tunnel...")

python_executable = sys.executable

# Get the site-packages directory for the current Python environment
site_packages_path = !{python_executable} -c "import site; print(site.getsitepackages()[0])"
site_packages_path = site_packages_path[0]

kohya_gui_path = "/content/gdrive/MyDrive/sd/kohya_ss_colab/kohya_gui.py"
base_shell_command = f"PYTHONPATH={site_packages_path}:$PYTHONPATH {python_executable} {kohya_gui_path}"

# Check if the user has provided a username and password
# Note: This relies on User and Password variables still being in the kernel from the previous run
if 'User' in locals() and 'Password' in locals() and User and Password:
    # Append username and password to the command, WITHOUT --share
    kohya_command = f"{base_shell_command} --username {User} --password {Password}"
else:
    # Run without credentials, WITHOUT --share
    kohya_command = f"{base_shell_command}"

def run_kohya_gui():
    # Execute the constructed full shell command for Kohya SS WebUI
    !{kohya_command}

# Start Kohya SS WebUI in a separate thread
kohya_thread = threading.Thread(target=run_kohya_gui)
kohya_thread.daemon = True # Allow the thread to exit if the main program exits
kohya_thread.start()

# Give Kohya some time to start up
time.sleep(15) # Adjust this if Kohya takes longer to start

try:
    # Open a ngrok tunnel to the Gradio port (default 7860)
    public_url = ngrok.connect(7860)
    print(f"\nKohya SS WebUI is accessible via ngrok at: {public_url}\n")
    print("This link will remain active as long as this cell is running.")
except Exception as e:
    print(f"Failed to create ngrok tunnel: {e}")
    print("Please ensure Kohya SS WebUI started correctly and your ngrok auth token is valid.")
````
